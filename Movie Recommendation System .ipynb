# User Interface with Feedback Mechanism

```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox
from tkinter import simpledialog
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
class MovieRecommendationApp:
    def __init__(self, master):
        self.master = master
        master.title("Movie Recommendation System")

        # Center align the window
        window_width = 400
        window_height = 200
        screen_width = master.winfo_screenwidth()
        screen_height = master.winfo_screenheight()
        x_coordinate = int((screen_width / 2) - (window_width / 2))
        y_coordinate = int((screen_height / 2) - (window_height / 2))
        master.geometry("{}x{}+{}+{}".format(window_width, window_height, x_coordinate, y_coordinate))

        
        self.label = tk.Label(master, text="Enter a movie title:", font=("Helvetica", 16, "bold"))
        self.label.pack(pady=10)

        self.entry = tk.Entry(master, font=("Helvetica", 14))
        self.entry.pack(pady=10)

        self.button = tk.Button(master, text="Get Recommendations", font=("Helvetica", 14, "bold"), command=self.get_recommendations)
        self.button.pack(pady=10)

    def get_recommendations(self):
        movie_title = self.entry.get()
        if movie_title:
            try:
                recommendations = self.get_movie_recommendations(movie_title)
                recommendation_text = "Recommendations based on '{}':\n".format(movie_title)
                for i, (movie, distance_val) in enumerate(recommendations, start=1):
                    recommendation_text += "{}. {} (distance: {})\n".format(i, movie, distance_val)
                feedback = messagebox.askyesno("Recommendations", recommendation_text + "\n\nWould you like to provide feedback?")
                if feedback:
                    self.save_feedback(movie_title)
            except KeyError:
                messagebox.showerror("Error", "Movie '{}' not found in the dataset.".format(movie_title))
            except Exception as e:
                messagebox.showerror("Error", str(e))
        else:
            messagebox.showerror("Error", "Please enter a movie title.")
    def save_feedback(self, movie_title):
        feedback = simpledialog.askstring("Feedback", "Please provide your feedback for '{}' (optional):".format(movie_title))
        if feedback:
            with open("feedback.txt", "a") as file:
                file.write("Movie: {}\nFeedback: {}\n\n".format(movie_title, feedback))
            messagebox.showinfo("Feedback", "Thank you for your feedback! Take care of you and yours:)")
        else:
            messagebox.showinfo("Feedback", "No feedback provided.")
    def get_movie_recommendations(self, movie_title):
        data = pd.read_csv('Dataset.csv')
        movies = pd.read_csv('Movie_Id_Titles.csv')
        df = pd.merge(data, movies, on='item_id')
        max_count_rate = df.groupby('title')['rating'].count().reset_index().rename(columns={'rating': 'total_rate_count'})
        df2 = pd.merge(df, max_count_rate, on='title')
        threshold = 116
        final = df2.query('total_rate_count >= @threshold')
        pivot_table = final.pivot_table(index='title', columns='user_id', values='rating').fillna(0)
        pivot_table_array = csr_matrix(pivot_table.values)
        model = NearestNeighbors(metric="cosine", algorithm="brute")
        model.fit(pivot_table_array)
        movie_index = pivot_table.index.get_loc(movie_title)
        distance, indices = model.kneighbors(pivot_table.iloc[movie_index, :].values.reshape(1, -1), n_neighbors=7)
        recommendations = []
        for i in range(1, len(indices.flatten())):
            movie = pivot_table.index[indices.flatten()[i]]
            distance_val = distance.flatten()[i]
            recommendations.append((movie, distance_val))

        return recommendations

def main():
    root = tk.Tk()
    app = MovieRecommendationApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
